```python
import pandas as pd
```

```python
df = pd.read_csv("../data/risk_data.csv")
df.head(3)
```

```python
all_labels = []
for label in df['labels']:
    a = label.split(",")
    for l in a:
        if l != "Ambiguous":
            all_labels.append(l)
```

```python
len(all_labels)
```

```python
(pd.Series(all_labels).value_counts()/len(all_labels)) * 100
```
